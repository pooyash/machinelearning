```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading data
df = pd.read_csv('penguins.csv')
df.tail(5)
```

```python
fig, ax = plt.subplots()
for name, group in df.groupby('species'):
    ax.scatter(group['culmen_length_mm'], group['culmen_depth_mm'], label=name)

plt.legend()
plt.xlabel('culmen_length_mm')
plt.ylabel('culmen_depth_mm')
plt.title('culmen_depth/culmen_length relation', loc='left');
```

```python
fig, ax = plt.subplots()
for name, group in df.groupby('species'):
    ax.scatter(group['culmen_length_mm'], group['flipper_length_mm'], label=name)

plt.legend()
plt.xlabel('culmen_length_mm')
plt.ylabel('flipper_length_mm')
plt.title('flipper_length/culmen_lengt relation', loc='left');
```

```python
fig, ax = plt.subplots()
for name, group in df.groupby('species'):
    ax.scatter(group['culmen_depth_mm'], group['body_mass_g'], label=name)

plt.legend()
plt.xlabel('culmen_length_mm')
plt.ylabel('body_mass_g')
plt.title('body_mass/culmen_lengt relation', loc='left');
```
