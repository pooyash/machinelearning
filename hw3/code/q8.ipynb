```python
import numpy as np
N = 1000
np.random.seed(1)
X = np.concatenate((np.random.normal(0,1,int(0.3*N)),
                    np.random.normal(5,1,int(0.7*N))))[:, np.newaxis]
```

```python
#importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.neighbors import NearestCentroid
```

```python
from scipy.fftpack import fft, fftshift
#parzen window
w = signal.parzen(len(X))
plt.plot(w , color = 'firebrick')
plt.title("Parzen window")
plt.ylabel("Amplitude")
plt.xlabel("Number of Sample")
```

```python
#defining functions for use in parzen window algotithm
import math
def gaussian(x):
    # mu = 0 and sigma = 1
    return math.exp(-x*x/2.0) / math.sqrt(2.0*math.pi)

def parzen_estimation(data, window_width):
    n = len(data)
    rst = []
    for x in data:
        a=0.0
        for xi in data:
            a+= gaussian((x-xi)/window_width)
        p_n = a/(n*window_width)
        rst.append(p_n)
    return rst
```

```python
#sorting data for using in plot
d= X
d_sorted = np.sort(d)
```

```python
#h=0.1
y_estimation= parzen_estimation(d_sorted, 0.1)
plt.figure(figsize=(10, 6))
plt.plot(d_sorted, y_estimation , color = 'salmon')
plt.title('Parzen Estimation for h=0.1')
plt.xlabel('Duration')
plt.ylabel('Probability')
```

```python
#h=1
y_estimation= parzen_estimation(d_sorted, 1)
plt.figure(figsize=(10, 6))
plt.plot(d_sorted, y_estimation , color = 'orchid')
plt.title('Parzen Estimation for h=0.1')
plt.xlabel('Duration')
plt.ylabel('Probability')
```
